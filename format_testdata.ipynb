```python
import pandas as pd
```

```python
lat_lon_dict = {'BKK':[13.729984,100.536443],
'Chiangmai':[18.840633,98.969661],
'Khonkaen':[16.445329,102.835251],
'Rayong':[12.671521,101.275875],
'Saraburi':[14.685833,100.871996],
'Surat':[9.126057,99.325355]}
```

# BKK

```python
bkk = pd.read_csv('./BKK_Scape/BKK_scape.csv')
bkk = bkk.iloc[:,1:]
bkk['date'] = pd.to_datetime(bkk['date'])
bkk = bkk[['date','PM2.5','temp','wind','dir']]
bkk.columns = ['date_time','PM2.5','temp','wind speed','wind dir']
bkk['lat'] = lat_lon_dict['BKK'][0]
bkk['long'] = lat_lon_dict['BKK'][1]

bkk.head()
```

```python
bkk.to_csv('BKK_clean.csv')
```

# Chiangmai

```python
chiangmai = pd.read_csv('./Chiangmai_Scape/Chiangmai_withTempAndWind0.csv')
chiangmai = chiangmai.iloc[:,1:]
chiangmai['date'] = pd.to_datetime(chiangmai['date'])
chiangmai.head()
```

```python
chiangmai_temp = pd.read_csv('./Chiangmai_Scape/Chiangmai_withTemp0.csv')
chiangmai_temp = chiangmai_temp.iloc[:,1:]
chiangmai_temp['date'] = pd.to_datetime(chiangmai_temp['date'])
chiangmai_temp.head()
```

```python
chiangmai['temp']=chiangmai_temp['temp']
chiangmai.head()
```

```python
chiangmai = chiangmai[['date','PM2.5','date_utc','url_date','temp','wind']]
chiangmai.head()
```

```python
for i in range(1,9):
    chiangmai2 = pd.read_csv('./Chiangmai_Scape/Chiangmai_withTempAndWind'+str(i)+'.csv')
    chiangmai2 = chiangmai2.iloc[:,1:]
    chiangmai = pd.concat([chiangmai,chiangmai2],ignore_index=True)
    
```

```python
chiangmai.tail(10)
```

```python
chiangmai['dir']=chiangmai['wind'].apply(lambda x: x.split("° @ ")[0])
chiangmai['wind']=chiangmai['wind'].apply(lambda x: x.split("° @ ")[1])
chiangmai.head()
```

```python
chiangmai = chiangmai[['date','PM2.5','temp','wind','dir']]
chiangmai.columns = ['date_time','PM2.5','temp','wind speed','wind dir']
chiangmai['lat'] = lat_lon_dict['Chiangmai'][0]
chiangmai['long'] = lat_lon_dict['Chiangmai'][1]
chiangmai.head()
```

```python
chiangmai.to_csv('Chiangmai_clean.csv')
```
