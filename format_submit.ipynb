```python
import time
import os
import pandas as pd
from datetime import datetime, timedelta
from tqdm.notebook import tqdm_notebook
tqdm_notebook.pandas()
```

```python
data = {'time': [], 'predicted': []}
start_time = datetime(2020,7,1,0,0,0)
end_time = datetime(2021,7,2,0,0,0)
```

```python
while start_time < end_time:
    pred_time = start_time
    for i in range(12):
        pred_time += timedelta(hours=6)
        data['time'].append(start_time)
        data['predicted'].append(pred_time)
    start_time += timedelta(hours=6)
```

```python
df = pd.DataFrame.from_dict(data)
```
