```python
lat_lon_dict = {'BKK':[13.729984,100.536443],'Chiangmai':[18.840633,98.969661],'Khonkaen':[16.445329,102.835251],'Rayong':[12.671521,101.275875],'Saraburi':[14.685833,100.871996],'Surat':[9.126057,99.325355]}
```

```python
def nearest_latlong(lat,long):

    # lat = row['latitude']
    # long = row['longitude']

    m_idx = 'BKK'
    m = (lat_lon_dict[m_idx][0]-lat)**2 + (lat_lon_dict[m_idx][1]-long)**2

    for key,value in lat_lon_dict.items():
        temp = (value[0]-lat)**2 + (value[1]-long)**2
        if(m>temp):
            m_idx = key
            m = temp
    
    if(m>0.5):
        return None
    return m_idx
    
```

```python
hotspot['province']=hotspot.apply(lambda x: nearest_latlong(x['latitude'],x['longitude']),axis=1)
hotspot.head()
```

```python
from datetime import timedelta,datetime
```

```python
def datetimeToFormatDateTime(date,time):
    time = int(time)
    y,m,d = date.split("-")
    minute = time%100
    hour = time//100

    dt = datetime(int(y),int(m),int(d),hour,minute)

    if(minute>=30):
        dt += timedelta(hours=1)
    dt -= timedelta(minutes=minute)

    return dt
        
```

```python
hotspot['date_time']=hotspot.apply(lambda x: datetimeToFormatDateTime(x['acq_date'],x['acq_time']),axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse
```

```python
# df_test = pd.read_csv('Chiangmai_clean.csv')
df_test = pd.read_csv('BKK_clean.csv')
df_test = df_test.iloc[:,1:]
df_test['date_time'] = pd.to_datetime(df_test['date_time'])
df_test.dropna(axis=1,inplace=True)
df_test.head()
```

```python
# df_train = pd.read_csv('Chiangmai/train/chiangmai_train_format.csv')
df_train = pd.read_csv('BKK/train/bkk_train_format.csv')
df_train['date_time'] = pd.to_datetime(df_train['date_time'])
df_train.dropna(axis=1,inplace=True)

df_train.head()
```

```python
df_train.isna().sum()
```

```python
train_X, train_y = df_train.loc[:,['temp','wind speed','wind dir']], df_train.loc[:,['PM2.5']]
test_X, test_y = df_test.loc[:,['temp','wind speed','wind dir']], df_test.loc[:,['PM2.5']]
```

```python
test_X.head()
```

```python
df_train.info()
```

```python
a = []
```

```python
for p in range(0,2):
    for q in range(0,2):
        best_order = (p, 0, q)


        model= SARIMAX(train_y, 
        exog=train_X,
        order=best_order,
        enforce_invertibility=False, enforce_stationarity=False)

        results = model.fit()

        predictions= results.predict(start =0, end=len(test_X)-1,exog=test_X)
        predictions=pd.DataFrame(predictions)
        predictions.reset_index(drop=True, inplace=True)
        predictions.index=test_X.index
        predictions['Actual'] = test_y['PM2.5']
        predictions.rename(columns={0:'Pred'}, inplace=True)
        predictions.columns = ['Pred','Actual']
        error=rmse(predictions['Pred'], predictions['Actual'])
        a.append({'best_order':best_order,'results':results,'error':error})

        print(best_order)
        print(error)

        
```

```python
for m in a:
    print(m['best_order'])
    print(m['error'])
```
