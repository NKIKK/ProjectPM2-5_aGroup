```python
import pandas as pd
```

```python
provinces = ['BKK','Chiangmai','Khonkaen','Rayong','Saraburi','Surat']
province = provinces[0]

data_test = pd.read_csv(province+'_clean.csv', date_parser = True)
data_test = data_test.drop(['Unnamed: 0', 'lat', 'long'], axis = 1)
data_test.head()
```

```python
def chk_time(x):
    
    if(x[11:13] in ['00','06','12','18']):
        return True
    return False
```

```python
data_test['chk_time'] = data_test['date_time'].apply(lambda x: chk_time(x))
data_test.head(10)
```

```python
data_test = data_test[data_test['chk_time']]
data_test.head(10)
```

```python
data_test.drop(['chk_time'],axis=1,inplace=True)
```

```python
data_test
```

```python
data_test.to_csv(province+'_clean2.csv',index=False)
```
